```python
import pandas as pd
import numpy as np
```

```python
# https://www.cs.toronto.edu/~mangas/teaching/320/slides/CSC320L11.pdf
# https://www.cse.iitd.ac.in/~pkalra/csl783/haar.pdf
def encode_haar(inputs):
  outputs = []
  while(len(inputs) > 1):
    elements = []
    haar_coefficient = []
    for i in range(0, len(inputs), 2):
      elements.append(np.average([inputs[i], inputs[i+1]]))
      haar_coefficient.insert(0, (inputs[i] - inputs[i+1]) / 2)
    for i in range(0, len(haar_coefficient)):
      outputs.insert(0, haar_coefficient[i])
    inputs = elements
    print(inputs)
    print(outputs)
  outputs.insert(0, inputs[0])
  return outputs
```

```python
#inputs = [9, 7, 3, 5];
inputs = [9, 7, 3, 5, 6, 10, 2, 6];
encode_haar(inputs)
```

```python
def decode_haar(inputs):
  outputs = []
  outputs.append(inputs[0])
  scale_index = 1
  len_inputs = len(inputs)
  while(scale_index < len_inputs):
    tmp = outputs
    outputs = []
    for i in range(0, scale_index):
      print(i)
      print(i+scale_index)
      outputs.append(tmp[i] + inputs[i+scale_index])
      outputs.append(tmp[i] - inputs[i+scale_index])
    scale_index = scale_index * 2
    print(outputs)
  return outputs
```

```python
inputs = [6.0, 0.0, 2.0, 2.0, 1.0, -1.0, -2.0, -2.0]
decode_haar(inputs)
```
