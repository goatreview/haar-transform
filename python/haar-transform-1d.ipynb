```python
import pandas as pd
import numpy as np
# https://www.cs.toronto.edu/~mangas/teaching/320/slides/CSC320L11.pdf
# https://www.cse.iitd.ac.in/~pkalra/csl783/haar.pdf
```

```python
def encode_haar(inputs):
    len_inputs = len(inputs)
    outputs = [0] * len_inputs
    scale_index = int(len_inputs / 2)
    while(scale_index >= 1):
        for i in range(0, scale_index):
            outputs[i] = (inputs[2*i] + inputs[2*i+1]) / 2
            outputs[i+scale_index] = (inputs[2*i] - inputs[2*i+1]) / 2
        inputs = outputs[:scale_index*2]
        scale_index = int(scale_index / 2)
    return outputs
```

```python
inputs = [9, 7, 3, 5, 6, 10, 2, 6];
encode_haar(inputs)
```

```python
def decode_haar(inputs):
    len_inputs = len(inputs)
    outputs = inputs.copy()
    scale_index = 1
    while(scale_index < len_inputs):
        for i in range(scale_index):
            even = int(outputs[i] + inputs[i + scale_index])
            odd = int(outputs[i] - inputs[i + scale_index])
            outputs[i * 2] = even
            outputs[i * 2 + 1] = odd
        scale_index = scale_index * 2
    return outputs
```

```python
inputs = [6.0, 0.0, 2.0, 2.0, 1.0, -1.0, -2.0, -2.0]
decode_haar(inputs)
```
